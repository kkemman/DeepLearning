```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
```

<h1>데이터 전처리</h1>

```python
# 1. 데이터 로드 및 전처리
df = pd.read_csv('C:/resources/DeepLearning/DSA_features.csv')
df
```

```python
#데이터셋의 activity가 standing과 lying으로 시작하는 애들만 남도록
df = df[(df['activity'] == 'standing') | (df['activity'].str.startswith('lying'))]
df
```

```python
#곁측치 확인
df.isnull().values.any()
df = df.dropna()
```

```python
df.loc[df['activity'] == 'standing', 'activity'] = 1
df.loc[df['activity'].str.startswith('lying', na=False), 'activity'] = 0
df
```

```python
df['activity'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

data = df.drop(['people'], axis=1)

# 스케일러 생성
#scaler = StandardScaler()

# 전체 컬럼을 스케일링
#data = scaler.fit_transform(data)
#data
```

```python
# 테스트 레이블을 남겨두는 이유는 나중에 정확도를 비교하기 위해서임
RANDOM_SEED = 42
X_train, X_test = train_test_split(data, test_size=0.2, random_state=RANDOM_SEED)
X_train = X_train[X_train.activity == 0]
X_train = X_train.drop(['activity'], axis=1)

y_test = X_test['activity']
X_test = X_test.drop(['activity'], axis=1)

X_train = X_train.values
X_test = X_test.values
X_train.shape
```

<h1>비선형 오토인코더</h1>

```python
# 오토인코더 모델

input_dim = X_train.shape[1]
encoding_dim = 14  # 인코딩 차원을 설정 (임의로 설정한 값, 조정 가능)

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
encoded = Dense(int(encoding_dim / 2), activation='relu')(encoded)
decoded = Dense(int(encoding_dim / 2), activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
```

```python
# 3. 모델 학습
# 훈련데이터의 0.1 을 validation data로 사용
history = autoencoder.fit(X_train, X_train,
                          epochs=100, 
                          batch_size=256, 
                          shuffle=True, 
                          validation_split = 0.1)
```

```python
# X_test 값을 넣어서 Reconstruction Error 값을 얻고,
# 실제 레이블인 y_test와 데이터프레임으로 구성
# Reconstruction Error + 실제 레이블의 형태가 됨

decoded_data  = autoencoder.predict(X_test)
reconstruction_error = np.mean(np.abs(X_test - decoded_data), axis=1)
# threshold 값을 조정하여 최선의 정확도를 얻음, 그래프로를 보고 결정

threshold = .83
predictions = (reconstruction_error > threshold).astype(int)
error_df = pd.DataFrame({'reconstruction_error': reconstruction_error,
                         'Predicted Label': predictions,
                         'true_class': y_test})
error_df.describe()
```

```python
error_df.head()
```

```python
groups = error_df.groupby('true_class')
fig, ax = plt.subplots()

for name, group in groups:
    ax.plot(group.index, group.reconstruction_error, marker='o', ms=3.5, linestyle='',
            label= "Fraud" if name == 1 else "Normal")
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Reconstruction error for different classes")
plt.ylabel("Reconstruction error")
plt.xlabel("Data point index")
plt.show();
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# true_class를 NumPy 배열로 변환
true_class = error_df.true_class.values

# y_pred를 NumPy 배열로 변환
y_pred = np.array([1 if e > threshold else 0 for e in error_df.reconstruction_error.values])

# 데이터 타입 확인
print("Data type of true_class:", type(true_class), true_class.dtype)
print("Data type of y_pred:", type(y_pred), y_pred.dtype)

# 정수형으로 변환
true_class = true_class.astype(int)
y_pred = y_pred.astype(int)

# 길이 확인
print("Length of true_class:", len(true_class))
print("Length of y_pred:", len(y_pred))

# 혼동 행렬 계산
conf_matrix = confusion_matrix(true_class, y_pred)

# 시각화
plt.figure(figsize=(8, 8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues')
plt.title("Confusion Matrix")
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.show()

# 분류 리포트 출력
print(classification_report(true_class, y_pred))
```
